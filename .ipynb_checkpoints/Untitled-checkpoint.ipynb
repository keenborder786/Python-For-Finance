```python
import random 
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel("PSX 100-returns and Betas.xlsx")
```

```python
df.set_index("Stocks",inplace=True)
df.head()
```

```python
overall_returns=[]
wieghts=1
for i in range(500):
    Trial_returns=[]
    for a,b in zip(df.Returns, df.Betas):
        exp=np.random.normal(a,abs(b))
        Trial_returns.append(exp*random.uniform(0, 1))
    returns=sum(Trial_returns)
    overall_returns.append(returns)
```

```python
import matplotlib.pyplot as plt
plt.plot(overall_returns)
plt.show()
```
