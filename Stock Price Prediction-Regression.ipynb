```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('data.xlsx', header=None)
df_2=pd.read_excel("data_2.xlsx",header=None)
df_3=pd.read_excel("data_3.xlsx",header=None)
```

```python
df.fillna(method="bfill",inplace=True)
df_2.fillna(method="bfill",inplace=True)
df_3.fillna(method="bfill",inplace=True)
```

```python
df_3.head()
```

```python
df.isnull().sum()
df_2.isnull().sum()
df_3.isnull().sum()
```

```python
df=df.merge(df_2,how="inner",left_index=True,right_index=True)
df=df.merge(df_3,how="inner",left_index=True,right_index=True)
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
column_names=df.columns.tolist()[1:]
final_df=pd.DataFrame()
```

```python
final_df= pd.melt(
    df, value_vars=column_names)
final_df
```

```python
final_df.drop("variable",1,inplace=True)
```

```python
final_df
```

```python
final_df.plot()
```

```python
final_df["label"]=final_df["value"].shift(-20)
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X=np.array(final_df.drop(["label"],1))
X=preprocessing.scale(X)
X_lately=X[-20:]
X=X[:-20]


final_df.dropna(inplace=True)
y=np.array(final_df["label"])
    
print(df)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)


clf=LinearRegression(n_jobs=-1)

clf.fit(X_train,y_train)
```

```python
accuracy=clf.score(X_test,y_test)
print(accuracy)
```

```python
forecast_set=clf.predict(X_lately)
print(forecast_set)
```
