```python
import numpy as np
import quandl
import matplotlib.pyplot as plt
```

```python
def gbm(mu,sigma,S,dt):
    drift=mu*S*dt
    violatility=sigma*S*((dt)**(1/2))*np.random.normal(0,dt)
    S1=drift+violatility+S
    return S1
Stock_prices=[]
for i in range(1000):
    S1=gbm(5,2,100,1/1000)
    Stock_prices.append(S1)
plt.plot(Stock_prices)
plt.show()
```

```python

def stock_price(mu,sigma,So,T):
    dt=1/T
    t=np.linspace(0,1,T)
    b = np.random.normal(0, 1, int(T))*np.sqrt(dt)  # brownian increments
    W = np.cumsum(b)                             # brownian path
#     print(W)
    S=[]
    S.append(So)
    for i in range(1,int(T)):
#         print(b)
        drift=(mu-(0.5*sigma**2))*t[i]
        violatility=sigma*W[i]
#         print(drift)
        S_temp=So*np.exp(drift+violatility)
#         print(S_temp)
        S.append(S_temp)
    return S,t
So = 100
mu = 0.15
sigma = 0.4
T = 1000

soln = stock_price(mu,sigma,So,T)[0]    # Exact solution
t = stock_price(mu,sigma,So,T)[1]       # time increments for  plotting 

plt.plot(t, soln)
plt.ylabel('Stock Price, $')
plt.title('Geometric Brownian Motion')
    
```
