```python
import random 
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel("PSX 100-returns and Betas.xlsx")
```

```python
df.set_index("Stocks",inplace=True)
df.head()
```

```python
overall_returns=[]
for i in range(500):
    Trial_returns=[]
    for a,b,c in zip(df.Returns, df.Betas,df.Weights):
        exp=np.random.normal(a,abs(b))
        Trial_returns.append(exp*c)
    returns=sum(Trial_returns)
    overall_returns.append(returns)
overall_returns
```

```python
import matplotlib.pyplot as plt
ax=plt.subplot(111)
ax.plot(overall_returns)
ax.set_xlabel("No.of trials")
ax.set_ylabel("Return of Portfolio")
plt.show()
```

```python
import seaborn as sns
sns.distplot(overall_returns)
```
